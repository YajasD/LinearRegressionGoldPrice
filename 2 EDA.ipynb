#### In this notebook, Gold price, CRB price, Inflation Rate, M1 Money Supply and EURUSD data is loaded and visually plotted against time to get an idea of what it looks like.

```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
% matplotlib inline
```

```python
##Import all datasets
gold_prices_df = pd.read_pickle('/Users/yajasdwivedi/metis/LinearRegressionPriceGold/gold_prices_df.pkl')
crb_df = pd.read_pickle('/Users/yajasdwivedi/metis/LinearRegressionPriceGold/crb_df.pkl')
inflation_rate_df_final = pd.read_pickle('/Users/yajasdwivedi/metis/LinearRegressionPriceGold/inflation_rate_df_final.pkl')
m1_df_final = pd.read_pickle('/Users/yajasdwivedi/metis/LinearRegressionPriceGold/m1_df_final.pkl')
eurusd_df = pd.read_pickle('/Users/yajasdwivedi/metis/LinearRegressionPriceGold/eurusd_df.pkl')
```

```python
##All features together##
plt.figure(dpi=150, figsize=(12,8))
##
plt.subplot(2,2,1)
plt.plot(crb_df['DATE'], crb_df['CRB_Price'])
plt.ylabel("CRB Index (in USD)")
plt.title("Price of CRB Index")
plt.xticks(rotation=45)
##
plt.subplot(2,2,2)
plt.plot(inflation_rate_df_final['DATE'], inflation_rate_df_final['inflation_rate_percent'])
plt.ylabel("Inflation Rate (%)")
plt.title('Inflation Rate')
plt.xticks(rotation=45)
##
plt.subplot(2,2,3)
plt.plot(m1_df_final['DATE'], (m1_df_final['M1SL']/1000))
plt.ylabel("M1 Money Supply (in Trillions of USD)")
plt.title('M1 Money Supply')
plt.xticks(rotation=45)
##
plt.subplot(2,2,4)
plt.plot(eurusd_df['DATE'], eurusd_df['EURUSD'])
plt.ylabel("EUR/USD")
plt.title('Cost of 1 EUR against USD')
plt.xticks(rotation=45)
##
plt.tight_layout()
```
